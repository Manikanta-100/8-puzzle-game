# **Pre-Processing**

```python
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for i in columns:
  plt.figure(1)
  plt.subplot(121)
  sns.distplot(data[i]);
  plt.subplot(122)
  data[i].plot.box(figsize=(16,5))
  plt.show()
```

```python
data.isnull().sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt

columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Setting up subplots
fig, axs = plt.subplots(1, len(columns), figsize=(15, 5))

for idx, col in enumerate(columns):
    percentile25 = data[col].quantile(0.25)
    percentile75 = data[col].quantile(0.75)

    upper_limit = percentile75 + 1.5 * (percentile75 - percentile25)
    lower_limit = percentile25 - 1.5 * (percentile75 - percentile25)

    # Capping
    data[col] = np.where(
        data[col] > upper_limit,
        upper_limit,
        np.where(
            data[col] < lower_limit,
            lower_limit,
            data[col]))

    # Plotting on respective subplot
    data[col].plot.box(ax=axs[idx])
    axs[idx].set_title(col)

plt.tight_layout()
plt.show()
```
